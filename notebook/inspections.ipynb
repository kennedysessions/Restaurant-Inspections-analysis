```python
# grouped the neighborhood tracking scores, averaged the scores for each neighborhood, and sorted them 
neighborhood_scores =dataframe.groupby("NTA")["SCORE"].mean().sort_values(ascending = False)
```

```python
# inspection scores by neighborhood tabulation area
neighborhood_scores.head(20)
```

```python
#neighborhood tabulation area and their average scores 
print(neighborhood_scores)
```

```python
#bottom 5 neighborhood inspection scores
neighborhood_scores.tail(5)
```

```python
#grouped all inspections by zipcode, averaged all the scores, and sorted from highest to lowest 
zipcode_scores = dataframe.groupby("ZIPCODE")["SCORE"].mean().sort_values(ascending = False)
```

```python
# top sanitation scores by zipcode
zipcode_scores.head(20)
```

```python
# bottom 20 sanitation scores by zipcodes 
zipcode_scores.tail(20)
```
